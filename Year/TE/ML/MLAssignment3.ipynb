```python
import pandas as pd 
import numpy as np
```

```python
ds = pd.read_csv("C:/Users/ASUS/Desktop/Machine Learning Lab/Datasets/Admission_Prediction.csv")
```

```python
ds
```

```python
ds1=ds.drop("Serial No.", axis=1)
```

```python
ds1.dtypes
```

```python
ds1.isnull().sum()
```

```python
ds1.mean(numeric_only=True)
```

```python
ds1.median(numeric_only=True)
```

```python
ds1.mode(numeric_only=True)
```

```python
from sklearn.preprocessing import Binarizer
```

```python
bi = Binarizer(threshold=0.88)
```

```python
ds1["Chance of Admit"]=bi.fit_transform(ds[["Chance of Admit"]])
```

```python
ds1
```

```python
x=np.array(ds1.drop("Chance of Admit", axis=1))
```

```python
x
```

```python
y=np.array(ds1["Chance of Admit"])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.75,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cl = DecisionTreeClassifier()
```

```python
cl.fit(x_train, y_train)
```

```python
y_pred=cl.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
matrix=confusion_matrix(y_test, y_pred)
```

```python
matrix
```

```python
report=classification_report(y_test, y_pred)
```

```python
print(report)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(cl)
```
