```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
ds=pd.read_csv("Market_Basket.csv",header=None)
ds
```

```python
ds.isna().sum()
```

```python
l1=ds.values.tolist()
l1
```

```python
nn=[]
for i in l1:
    row=[]
    for j in i:
        if not pd.isna(j):
            row.append(j)
    nn.append(row)
    
nn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ds.columns = ds.columns.astype(str)

col = '0'  


t = ds[col].value_counts().head(20).index


plt.figure(figsize=(8,5))
sns.countplot(x=col, data=ds[ds[col].isin(t)], order=t, )

plt.title('Frequency Plot')
plt.xlabel(col)
plt.ylabel('Count')
plt.show()
```

```python
from apyori import apriori
rules=apriori(nn,min_support=0.005,min_confidence=0.5)
rl=list(rules)
rl
```

```python
df1=pd.DataFrame()
rules1=[]
support1=[]
lift1=[]
a=[]
for i in rl1:
    rules1.append(i[2][0][0])
    a.append(i[2][0][1])
    support1.append(i[1])
    lift1.append(i[2][0][3])
    
df1['X']=rules1
df1['Y']=a
df1['Support']=support1
df1['Lift']=lift1

df1
```

```python
rules=apriori(nn,min_support=0.01,min_confidence=0.4)
rl2=list(rules)
rl2
```

```python
ds2=pd.DataFrame()
rules2=[]
support2=[]
lift2=[]

for i in rl2:
    rules2.append(i[0])
    support2.append(i[1])
    lift2.append(i[2][0][3])
    
ds2['Rules']=rules2
ds2['Support']=support2
ds2['lift']=lift2

ds2
```
